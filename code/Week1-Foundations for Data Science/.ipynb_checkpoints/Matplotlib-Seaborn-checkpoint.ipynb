### Swarm Plot

## Pairplot
- takes only numeric fields from DataFrame and plots them against each other, nXn plots
- diagonal line is against the same field so insted of showing 45 degree line it shows histogram showing frequency distribution of values
- All plots on one side of diagonal is mirror image(x & y flipped) of corresponding plots on the other side.
- number of unique plots are diagonal and all plots on one side of diagonal


```python
sns.pairplot(tips) 
plt.show()
```

```python
# sns.pairplot(data=tips, x='total_bill', y='tip', z='size') - wrong syntax
sns.scatterplot(tips[['total_bill', 'tip', 'size']])

plt.show()
```

### displot 
- histogram plus distribution density curve
- Asses distribution of sample data by comparing empirical distribution with theoritical values expected from specified distribution
- Distribution Plot is NOT a bivariate graphical representation, its univariate
- Distribution plot take into consideration the numerical data as primary input
- Distribution plot helps to understand a data better from the statistical perspective.
- Distribution Plot is an estimate of the population based on the data.
- Using a loop and different labels we can plot multiple distribution plot in a single graph


```python
sns.displot(tips['total_bill'])
plt.show()
```

```python
sns.displot(tips['total_bill'], kind='kde')
plt.show()
```

```python
time =  pd.Series(tips['time'].value_counts().keys())

for i in time:
    subset = tips[tips['time'] == i]
    
    sns.distplot(subset['total_bill'], 
                 hist=False, 
                 kde = True, 
                 kde_kws = {'shade':True},
                 label = i)
```
